#### Ejercicio 5: ¿Cuántos títulos se añadieron a Netflix cada año? Para resolver este ejercicio tendrás uqe convertir la columna `date_added` a formato fecha usando el método `pd.to_datetime`. [Aquí](https://pandas.pydata.org/docs/reference/api/pandas.to_datetime.html) tienes la documentación, leeala con detenimiento antes de usar el método.

#### Ejercicio 6: ¿Cuál es el país con más títulos lanzados en el año 2020? Para resolver este ejercicio después del gropuby os puede resultar interesante el método [`idxmax`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.idxmax.html)

#### Ejercicio 10. ¿Cuál es el promedio de duración de las películas agrupadas por país?

```python
df_type_movie = df_netflix_final[df_netflix_final["Type"] == "Movie"]
df_type_movie.sample()
```

```python
#Me voy a guardar en un dataframe las peliculas cuya duracion es nula por si acaso 
df_peliculas_sin_duracion = df_type_movie[df_type_movie["Duration"].isnull()]
df_peliculas_sin_duracion.sample()
```

```python
#Y voy a trabajar con estas que son las que realmente interesan 
df_peliculas_con_duracion = df_type_movie[df_type_movie["Duration"].notnull()]
df_peliculas_con_duracion.sample()
```

```python
df_peliculas_con_duracion["Duration_num"] = [duracion.split(" ")[0] for duracion in df_peliculas_con_duracion["Duration"]]
df_peliculas_con_duracion.sample()
```

```python
df_peliculas_con_duracion["Duration_num"] = df_peliculas_con_duracion["Duration_num"].astype(int)
```

```python
df_duracion_peliculas = round(df_peliculas_con_duracion.groupby("Country")["Duration_num"].mean(),2)
df_duracion_peliculas = df_duracion_peliculas.reset_index()
df_duracion_peliculas
```

#### Ejercicio 15: ¿Cuál es el lenguaje más frecuente en los títulos lanzados en los últimos 5 años agrupado por país?

```python
df_netflix_final["Release_year"].unique()
```

```python
df_lanzandos_ultimos = df_netflix_final[df_netflix_final["Release_year"] >= 2017]

df_lenguaje = df_lanzandos_ultimos.groupby(["Country", "Language"])["Show_id"].count()
df_lenguaje = df_lenguaje.reset_index()
df_lenguaje
```

```python
#Me he ayudado de varias paginas como: https://saturncloud.io/blog/how-to-select-row-with-max-value-in-column-from-pandas-groupby-groups/ 
#O esta: https://dareplanet.tech/insights/analisis-de-datos-libreria-pandas-python/
df_indices_lenguajes_frecuentes = df_lenguaje.groupby("Country")["Show_id"].idxmax()
df_lenguajes_frecuentes_pais = df_lenguaje.loc[df_indices_lenguajes_frecuentes]
df_lenguajes_frecuentes_pais
```
